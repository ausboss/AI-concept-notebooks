```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
# model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Predefined list of passages
passages = [
    "Merlin, played by Colin Morgan, has striking blue eyes and dark hair.",
    "Merlin is typically seen wearing a blue tunic and brown trousers, a typical attire for a servant in Camelot.",
    "He has a lean build and is of average height.",
    "Merlin possesses a kind and compassionate personality, often putting others' needs before his own.",
    "He is incredibly loyal, especially to his close friends Arthur, Gaius, and Gwen.",
    "Despite his gentle demeanor, Merlin is also brave and isn't afraid to stand up for what he believes in.",
    "Merlin is known for his quick wit and sense of humor.",
    "He is also very intelligent, with a particular talent for strategic thinking.",
    "Merlin holds a deep-seated belief in equality, which often puts him at odds with the strict hierarchy of Camelot.",
    "One of Merlin's most distinctive characteristics is his magical ability, which he must keep secret due to the laws of Camelot."
]
```

# Function to find the most related passage

```python
# Function to find the most related passage
def find_most_related(user_input):
    # Encode the user input and the passages
    user_input_embedding = model.encode(user_input)
    passage_embeddings = model.encode(passages)

    # Initialize the max score and most related passage
    max_score = -1
    most_related_passage = ''

    # Loop through each passage
    for passage, passage_embedding in zip(passages, passage_embeddings):
        # Calculate the similarity score
        score = util.dot_score(user_input_embedding, passage_embedding)

        # If this score is higher than the max score, update the max score and most related passage
        if score > max_score:
            max_score = score
            most_related_passage = passage

    # Return the most related passage and the max score
    return most_related_passage, max_score
```

# Enter a query
(i.e. What is the population of London?)

```python
# Get user input
user_input = input('Enter your query: ')

# Get the most related passage and the max score
most_related_passage, max_score = find_most_related(user_input)

# Print the most related passage and the max score
print('Most related passage:', most_related_passage)
print('Similarity:', max_score)
```
